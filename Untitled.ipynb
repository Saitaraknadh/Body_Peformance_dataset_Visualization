```python
# A short table showing the relationship between two or more variables
pd.crosstab(df["class"], df.gender)
```

```python
#Plotting a bar graph for the crosstab with a customized colour and size with names
pd.crosstab(df["class"], df.gender).plot(kind="bar", 
                                    figsize=(10,6), 
                                    color=["green", "lightgreen"]);
plt.title("Gender distribution by class")
plt.xlabel("A = Best, B = Well, C = Passable , D = Worst ")
plt.ylabel("Amount")
plt.legend(["Female", "Male"])
```

```python
sns.scatterplot(x='weight_kg',y='height_cm',data=df,hue='class',palette='crest')
plt.xlim(20,140)
plt.ylim(125,200)
plt.show()
```

```python
sns.scatterplot(x='gripForce',y='height_cm',data=df,hue='class',palette='crest')
plt.xlim(0,70)
plt.ylim(125,200)
plt.show()
```

```python
fig, axes = plt.subplots(1,4, figsize=(20,6))
var = ['gripForce', 'sit and bend forward_cm', 'sit-ups counts', 'broad jump_cm']
for var, axis in zip(var, axes):
    sns.boxplot(x='gender', y=var, data=df, palette='crest',ax=axis, hue='class')
axes[1].set_ylim(-30, 50);
```
